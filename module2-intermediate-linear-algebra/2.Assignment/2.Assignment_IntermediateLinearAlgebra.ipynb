## 1.1 Sales for the past week was the following amounts: [3505, 2400, 3027, 2798, 3700, 3250, 2689]. Without using library functions, what is the mean, variance, and standard deviation of of sales from last week? (for extra bonus points, write your own function that can calculate these two values for any sized list)

## Mean, Variance, Standard Deviation Functions.

```
# create a list of the values from above.
sales_list = [3505, 2400, 3027, 2798, 3700, 3250, 2689]
# create a function to calculate the 'mean'.
def mean(sales_list):
  return sum(sales_list)/len(sales_list)
# create a function to calculate the 'variance'.
def variance(sales_list):
  return sum([((entry-mean(sales_list))**2) for entry in sales_list])/len(sales_list)
# create a function to calculate the 'standard deviation'.
def stdv(sales_list):
  return (variance(sales_list)**.5)
# show the results.
print("sales list python 'mean' =", mean(sales_list))
print("sales list python 'variance' =", variance(sales_list))
print("sales list python 'standard deviation' =", stdv(sales_list))
```

## Numpy Mean, Variance, Standard Deviation.

```
# import numpy library.
import numpy as np
# create a list of the values from above.
sales_list = [3505, 2400, 3027, 2798, 3700, 3250, 2689]
numpy_mean = np.mean(sales_list)
numpy_var = np.var(sales_list)
numpy_std = np.std(sales_list)
print("sales list numpy 'mean' =", numpy_mean)
print("sales list numpy 'variance' =", numpy_var)
print("sales list numpy 'standard deviation' =", numpy_std)
```

## 1.2 Find the covariance between last week's sales numbers and the number of customers that entered the store last week: [127, 80, 105, 92, 120, 115, 93] (you may use librray functions for calculating the covariance since we didn't specifically talk about its formula)

## Covariance Function.

```
# label the sales & customers lists.
a = sales_list = [3505, 2400, 3027, 2798, 3700, 3250, 2689]
b = customers_list = [127, 80, 105, 92, 120, 115, 93] 
# create a function to find the 'covariance'.
def cov(sales_list, customers_list):
    if len(a) != len(b):
        return
    a_mean = np.mean(a)
    b_mean = np.mean(b)
    sum = 0
    for i in range(0, len(a)):
        sum += ((a[i] - a_mean) * (b[i] - b_mean))
    return sum/(len(a)-1)
# show the results.
print("sales list & customers list 'covariance' =", cov(sales_list, customers_list))
```

## 5.1 What is the span of the following vectors?

\begin{align}
g = \begin{bmatrix} 1 & 2 \end{bmatrix}
\qquad
h = \begin{bmatrix} 4 & 8 \end{bmatrix}
\end{align}

```
# create the 2 vectors above.
g = [1, 2]
h = [4, 8]
# create a function to return the 'span'.
def span(g):
    return max(g) - min(g)
def span(h):
    return max(h) - min(h)
# show the results.  
print("g =", g)  
print("g 'span' =", span(g))
print("h =", h)  
print("h 'span' =", span(h))
         
```

## 5.2 What is the span of $\{l, m, n\}$?

\begin{align}
l = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}
\qquad
m = \begin{bmatrix} -1 & 0 & 7 \end{bmatrix}
\qquad
n = \begin{bmatrix} 4 & 8  & 2\end{bmatrix}
\end{align}

```
# create the 3 vectors above.
l = [1, 2, 3]
m = [-1, 0, 7]
n = [4, 8, 2]
# multiplt l & m.
lm = np.multiply(l,m)
# multiplt lm & n.
lmn = np.multiply(lm, n)
# create a function to find the 'span'.
def span(lmn):
    return max(lmn) - min(lmn)
# show the results.
print("l =", l)
print("m =", m)
print("n =", n)
print("l*m =", lm)
print("lm*n =", lmn)
print("lmn 'span' =", span(lmn))
```
