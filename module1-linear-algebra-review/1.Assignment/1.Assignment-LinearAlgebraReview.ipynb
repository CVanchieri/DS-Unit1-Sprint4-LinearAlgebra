## Determinant Matrix.

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# create the 2 matrices.
H = np.array([[12, 11], [7, 10]])
J = np.array([[0, 1, 2], [7, 10, 4], [3, 2, 0]])
# create the determinents.
HD = np.linalg.det(H)
JD = np.linalg.det(J)
# show the results.
print("H =")
print(H)
print("HD =")
print(HD)
print("J =")
print(J)
print("JD =")
print(JD)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
# create the inverse matrices.
H1 = np.linalg.inv(H)
J1 = np.linalg.inv(J)
print("H =")
print(H)
print("H1 =")
print(H1)
print("J =")
print(J)
print("J1 =")
print(J1)
```
